```python
import numpy as np
import pandas as pd
```

```python
teams_df = pd.read_csv('data/mens_prem_teams.csv')
teams_df
```

```python
scores_df = pd.read_csv('data/2025-06-28_mens_prem_results.csv')
scores_df
```

```python
teams_list = teams_df.to_numpy()

def findTeamIndex(team_id, teams):
  for index, team in enumerate(teams):
    if team[0] == team_id:
      return index

def buildScoresArray(matches):
  rows = []
  for match in matches:
    row = [0] * len(teams_df.values)
    if match[1] >= match[2]:
      row[findTeamIndex(match[3], teams_list)] = 1
      row[findTeamIndex(match[4], teams_list)] = -1
      rows.append(row)
    else:
      row[findTeamIndex(match[4], teams_list)] = 1
      row[findTeamIndex(match[3], teams_list)] = -1
      rows.append(row)
  return rows

X = np.array(buildScoresArray(scores_df.to_numpy()))
X
```

```python
teams = teams_df['Club'].tolist()
match_ids = scores_df['Match ID'].tolist()

X_df = pd.DataFrame(X, index=match_ids, columns=teams)
print(X_df.to_string())
```

```python
print(X_df.T.to_string())
```

```python
M = (X.T).dot(X)
M_df = pd.DataFrame(M, index=teams, columns=teams)
print(M_df.to_string())
```

```python
# build GD array

def buildGDArray(matches, teams):
  gd_list = []
  for team in teams:
    gd = 0
    for match in matches:
      if team[0] == match[3]:
        gd += (match[1] - match[2])
      if team[0] == match[4]:
        gd += (match[2] - match[1])
    gd_list.append([gd])
  return gd_list

p = np.array(buildGDArray(scores_df.to_numpy(), teams_df.to_numpy()))
p
```

```python
M_bar = M.copy()
M_bar[-1,:] = np.ones(M.shape[0])
M_bar
M_bar_df = pd.DataFrame(M_bar, index=teams, columns=teams)
print(M_bar_df.to_string())
```

```python
p_bar = p
p_bar[-1] = 0
p_bar
p_bar_df = pd.DataFrame(p_bar, index=teams, columns=['GD'])
print(p_bar_df.to_string())
```

```python
r = np.linalg.inv(M_bar).dot(p_bar)
r
```

```python
r_df = pd.DataFrame(r, index=teams, columns=['Massey Rating']).sort_values(by=['Massey Rating'], ascending=False)
print(r_df)
```
